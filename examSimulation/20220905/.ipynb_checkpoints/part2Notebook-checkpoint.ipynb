# 2.

Algorithm for findPairs deliberately taken from [stack overflow](https://stackoverflow.com/questions/67954609/how-to-get-the-sum-of-elements-pairs-from-list)

```python
#class to implement point 2
class pairSum:
    s = 0
    int_list =[]
    
    def __init__(self,s=s,int_list=int_list):
        self.s = s
        self.int_list = int_list
        
    def findPairs(self):
        pairs = []
        indices = {x: i for i, x in enumerate(self.int_list)}
        for i, x in enumerate(self.int_list):
            target = self.s - x
            #here the first conditions find the pair of element that sums to s
            #the second removes pair in which there one element repetead twice
            #the third removes permutations of the same pair
            if target in indices and i!=indices[target] and (self.int_list[indices[target]],x) not in pairs:
                pairs.append( (x, self.int_list[indices[target]]) )
        return pairs

#fast check of correct behavior
test = pairSum(12,[2, 4, 5, 6, 7, 8])
test.findPairs()
test.s=13
test.findPairs()
```

```python
#generation of N random integer between min and max
minInt = 1
maxInt = 100
N = 100 
randList = np.random.randint(minInt,maxInt,N)
randList

#test of the class on random integers:
testSum = 128
randTest = pairSum(testSum,randList)
pair = randTest.findPairs()
pair
check = [a+b for a,b in pair]
check
```

# 3.

```python
#class to implement point 2
class zeroTriplets:
    int_list =[]
    
    def __init__(self,int_list=int_list):
        self.int_list = int_list
        
    def findTriplets(self):
        triplets = []
        indices = {x: i for i, x in enumerate(self.int_list)}
        for i, x in enumerate(self.int_list):
            targetTriplets =  - x              #for each element i implement a pair find with this sum
            
            pairs = []
            for j, y in enumerate(self.int_list):
                targetPairs = targetTriplets - y
                #here the first conditions find the pair of element that sums to s
                #the second removes pair in which there one element repetead twice
                #the third removes permutations of the same pair
                if targetPairs in indices and j!=indices[targetPairs] and (self.int_list[indices[targetPairs]],y) not in pairs:
                    pairs.append( (y, self.int_list[indices[targetPairs]]) )
                    newTriplet =  [x,y,self.int_list[indices[targetPairs]]] #I sort to avoid permutations
                    newTriplet.sort() 
                    if newTriplet not in triplets and i!=j and i!=indices[targetPairs]:
                        triplets.append( newTriplet ) #I add the element and the corresponding pair together 
        
            
        return triplets
            

#fast check of correct behavior
int_list = range(-10,10,2)
test = zeroTriplets(int_list)
triplets = test.findTriplets()
triplets
check = [a+b+c for a,b,c in triplets]
check
```

```python
#derived class
class pairs_and_triplets(pairSum,zeroTriplets):
    def __init__(self,s,int_List):
        super().__init__(s,int_list)
        self.int_list = int_list

#fast check of correct behavior
mySum = 6
int_list = range(-10,10,2)
finalClass = pairs_and_triplets(mySum,int_list)  
finalClass.findPairs()
finalClass.findTriplets()
```
