```python
#import relevant stuff
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize as opt
import emcee

#with these we can print variables 'inline'
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

# 1.

```python
#load relevant data
infile='true.dat'
data = np.asarray( np.loadtxt(infile) ) 
data

#gaussian displacement
mu=0
sig=0.5 * 10**(-12) # [F/m]

offdata = data + np.random.normal(mu,sig,size=np.shape(data))
data

#saving modified data to outputfile
outfile = 'offset.dat'
np.savetxt(outfile, data, delimiter="\t")
```

# 2.

```python
#Plot of data
%matplotlib inline
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(15,6))

ax1.grid()

n_bins=100
_ = ax1.hist(data[:,0], bins=n_bins, label=r'$C_{1}$ (no offset)', 
                           color='red', histtype='stepfilled', lw=2,alpha=0.6)
_ = ax1.hist(offdata[:,0], bins=n_bins, label=r'$\tilde C_{1}$ (offset)', 
                           color='blue', histtype='step', lw=2,alpha=0.5)

_ = ax1.set_title(r'Distribution of $C_1$')
_ = ax1.set_xlabel(r'$C_1$ [F/m]')
_ = ax1.set_ylabel('Occurence')

_ = ax1.legend()

ax2.grid()

n_bins=100
_ = ax2.hist(data[:,1], bins=n_bins, label=r'$C_{eq}$ (no offset)', 
                           color='red', histtype='stepfilled', lw=2,alpha=0.6)
_ = ax2.hist(offdata[:,1], bins=n_bins, label=r'$\tilde C_{eq}$ (offset)', 
                           color='blue', histtype='step', lw=2,alpha=0.5)

_ = ax2.set_title(r'Distribution of $C_{eq}$')
_ = ax2.set_xlabel(r'$C_{eq}$ [F/m]')
_ = ax2.set_ylabel('Occurence')

_ = ax2.legend()
```

```python
#get data for the fit
nBins = 100
binned_data_1, bins_1 = np.histogram(offdata[:,0], bins=nBins)
centerBins_1 = 0.5*(bins_1[1:] + bins_1[:-1])
binned_data_eq, bins_eq = np.histogram(offdata[:,1], bins=nBins)
centerBins_eq = 0.5*(bins_eq[1:] + bins_eq[:-1])
```

```python
#fit
def fit_func(x,a,b):
    return 1/( 1/(x+a) +b )

#for C1 with offset
pars_1, pars_cov_1 = opt.curve_fit(fit_func,centerBins_1,binned_data_1)
perr_1 = np.sqrt(np.diag(pars_cov_1))
#for Ceq with offset
pars_eq, pars_cov_eq = opt.curve_fit(fit_func,centerBins_eq,binned_data_eq)
perr_eq = np.sqrt(np.diag(pars_cov_eq))
```

```python
#Plot of data +fit
%matplotlib inline
fig, (ax1,ax2) = plt.subplots(2,1,figsize=(11,13))

ax1.grid()

n_bins=100
_ = ax1.hist(data[:,0], bins=n_bins, label=r'$C_{1}$ (no offset)', 
                           color='red', histtype='stepfilled', lw=2,alpha=0.6)
values, *_ = ax1.hist(offdata[:,0], bins=n_bins, label=r'$\tilde C_{1}$ (offset)', 
                           color='blue', histtype='step', lw=2,alpha=0.5)

_ = ax1.set_title(r'Distribution of $C_1$ + fit')
_ = ax1.set_xlabel(r'$C_1$ [F/m]')
_ = ax1.set_ylabel('Occurence')


x=np.linspace(np.min(offdata[:,0]), np.max(offdata[:,0]))
_ = ax1.plot(x,fit_func(x,pars_1[0],pars_1[1]),label=r'Fit over $\tilde C_{1}$', color='blue', lw=3)


_ = ax1.legend()

#Display text box with fitparameters outside plot
textstr = '\n'.join((
    'Fit parameters:',
    r'$a=%.2f \pm %.2f$' % (pars_1[0], perr_1[0] ),
    r'$b=%.2f \pm %.2f$' % (pars_1[1], perr_1[1], ),  ))

props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
# place the text box in upper left in axes coords
_ = ax1.text(1.05, 0.95, textstr, transform=ax1.transAxes, fontsize=14,
        verticalalignment='top', bbox=props)

ax2.grid()

n_bins=100
_ = ax2.hist(data[:,1], bins=n_bins, label=r'$C_{eq}$ (no offset)', 
                           color='red', histtype='stepfilled', lw=2,alpha=0.6)
_ = ax2.hist(offdata[:,1], bins=n_bins, label=r'$\tilde C_{eq}$ (offset)', 
                           color='blue', histtype='step', lw=2,alpha=0.5)

_ = ax2.set_title(r'Distribution of $C_{eq}$ + fit')
_ = ax2.set_xlabel(r'$C_{eq}$ [F/m]')
_ = ax2.set_ylabel('Occurence')

x=np.linspace(np.min(offdata[:,1]), np.max(offdata[:,1]))
_ = ax2.plot(x,fit_func(x,pars_eq[0],pars_eq[1]),label=r'Fit over $\tilde C_{eq}$', color='blue', lw=3)

_ = ax2.legend()

#Display text box with fitparameters outside plot
textstr = '\n'.join((
    'Fit parameters:',
    r'$a=%.2f \pm %.2f$' % (pars_eq[0], perr_eq[0] ),
    r'$b=%.2f \pm %.2f$' % (pars_eq[1], perr_eq[1], ),  ))

props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
# place the text box in upper left in axes coords
_ = ax2.text(1.05, 0.95, textstr, transform=ax2.transAxes, fontsize=14,
        verticalalignment='top', bbox=props)
```

# 3.

```python
#reading data
Ceq = data[:,1]
Ceq
#generating errors
mu_deltaCeq = 0
sigma_deltaCeq = 3 *10**(-12)
deltaCeq = np.random.normal(mu_deltaCeq, sigma_deltaCeq,size=np.shape(Ceq))
deltaCeq
#storing errors
#saving modified data to outputfile
outfile = 'errorbar.dat'
np.savetxt(outfile, data, delimiter="\t")
```

# 4.

```python
C1t = offdata[:,0]
```

```python
truths = np.asarray( [ 10*10**(-12), 1/( 20 *10**(-12) ) ] )
truths
```

```python
def log_likelihood(theta, x, y, yerr):
    a,b = theta
    model = fit_func(x,a,b)
    sigma2 = yerr ** 2 
    return -0.5 * np.sum((y - model) ** 2 / sigma2 + np.log(sigma2))
```

```python
#setting up monte carlo simulation
def log_prior(theta):
    a,b = theta
    # Return a finite value if theta is in the parameter space region with support for the prior
    if 0.5*truths[0] < a < 1.5*truths[0] and 0.5*truths[1] < b < 1.5*truths[1]:
        return 0.0
    # Otherwise return -infinity, basically a massive down-weight for the posterior
    return -np.inf

def log_probability(theta, x, y, yerr):
    lp = log_prior(theta)
    # If theta is outside the support for the prior, return -infinity
    if not np.isfinite(lp):
        return -np.inf
    # Otherwise return the logarithm of prior*likelihood
    return lp + log_likelihood(theta, x, y, yerr)
```

```python
#!pip install tqdm
```

```python
# The number of walkers
nwalkers = 32

# The dimensionality of the parameter space (a,b)
ndim = 2

# The MCMC steps
nmcmc = 5000

# Initial position
#pos = truths + 1e-11 * np.random.randn(nwalkers, ndim)
pos = np.asarray( [ [truths[0],truths[1]] for i in range(nwalkers)] )
pos[:,0] += 0.1*truths[0] * np.random.randn(nwalkers) 
pos[:,1] += 0.1*truths[1] * np.random.randn(nwalkers)


# Run the sampler
sampler = emcee.EnsembleSampler(nwalkers, ndim, log_probability, args=(C1t, Ceq, deltaCeq))
sampler.run_mcmc(pos, nmcmc, progress=True); # The ; avoids some prints to screen
```

```python
fig, axes = plt.subplots(2, figsize=(10, 7), sharex=True)
samples = sampler.get_chain()
labels = ["a", "b"]
col = ['r', 'g']
for i in range(ndim):
    ax = axes[i]
    ax.plot(samples[:, :, i], col[i], alpha=0.2) # 
    ax.set_xlim(0, len(samples))
    ax.set_ylabel(labels[i])
    ax.yaxis.set_label_coords(-0.1, 0.5)

axes[-1].set_xlabel("MCMC step"); # The ; avoids an unwanted output to screen.
```

```python
tau = sampler.get_autocorr_time()
print(tau)
```

```python
flat_samples = sampler.get_chain(discard=100, thin=20, flat=True)
print(flat_samples.shape)
```

```python
mc_fit_a = np.mean(flat_samples[:,0])
mc_fit_b = np.mean(flat_samples[:,1])
mc_fit_a
mc_fit_b
```

```python
#Plot of data +fit
%matplotlib inline
fig, ax = plt.subplots()

ax.grid()

ax.errorbar(C1t, Ceq, yerr=deltaCeq, fmt=".k", capsize=0)

_ = ax.set_title(r' $C_eq$ vs $\tilde C_1$  + fit')
_ = ax.set_xlabel(r'$C_1$ [F/m]')
_ = ax.set_ylabel('$C_eq [F/m]$')


x=np.linspace(np.min(C1t), np.max(C1t))
_ = ax.plot(x,fit_func(x,mc_fit_a,mc_fit_b),label=r'Fit', color='blue', lw=3)
_ = ax.plot(x,fit_func(x,truths[0],truths[1]),label=r'True', color='red', lw=3)


_ = ax.legend()

#Display text box with fitparameters outside plot
textstr = '\n'.join((
    'Fit parameters:',
    r'$a=%.2f' % (mc_fit_a ),
    r'$b=%.2f' % (mc_fit_b ),  ))

props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
# place the text box in upper left in axes coords
_ = ax1.text(1.05, 0.95, textstr, transform=ax1.transAxes, fontsize=14,
        verticalalignment='top', bbox=props)

```
