```python
#import relevant libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize as opt  
```

# 1.

```python
N = 10**4 #value to generate
storeDict = {"m1":[],"chi1":[],"m2":[],"chi2":[],"theta1":[],"theta2":[]}

m_min = 10
m_max = 100
chi_min = 0
chi_max = 1
cos_min = 0
cos_max = 1

storeDict["m1"] = np.random.uniform(m_min,m_max,N)
storeDict["m2"] = np.random.uniform(m_min,m_max,N)
storeDict["chi1"] = np.random.uniform(chi_min,chi_max,N)
storeDict["chi2"] = np.random.uniform(chi_min,chi_max,N)
storeDict["theta1"] = np.arccos( np.random.uniform(cos_min,cos_max,N) )
storeDict["theta2"] = np.arccos( np.random.uniform(cos_min,cos_max,N) )
storeDict
```

# 2.

```python
storeDict2 = {"chi_eff":[],"M":[],"eta":[],"M_chirp":[],"tc":[]}
storeDict2["M"] = storeDict["m1"] + storeDict["m2"]
storeDict2["chi_eff"] =  ( storeDict["m1"]*storeDict["chi1"]*np.cos(storeDict["theta1"]) + storeDict["m2"]*storeDict["chi2"]*np.cos(storeDict["theta2"]) ) / storeDict2["M"]
storeDict2["eta"] = 0.25 * storeDict["m1"] * storeDict["m2"] / storeDict2["M"]**2
storeDict2["M_chirp"] = (storeDict["m1"] * storeDict["m2"])**(3/5) / storeDict2["M"]**(1/5)
nu10 = 4.916*10**(-5)
r0 = (storeDict2["M"]/(np.pi*nu10)**2)**(1/3)
storeDict2["tc"] = 50/256 * nu10 * r0**4 /( storeDict["m1"]*storeDict["m2"]*storeDict2["M"])
storeDict2
```

# 3.

```python
list(storeDict2.keys())
```

```python
# Plot of distribution
fig, ax = plt.subplots(2,3,figsize=(15,15))

keys = list(storeDict2.keys() )
values = list(storeDict2.values() )
units = ["","Solar masses","","Solar Masses","s"]

for i,axi in enumerate(ax.flatten()):
    if i == len(keys): break
    axi.grid()
    _  = axi.hist(values[i],bins=100,color='Blue')
    _ = axi.set_title(f'Distribution of {keys[i]}')
    _ = axi.set_xlabel(units[i])
    _ = axi.set_ylabel("Occurences")
```

# 4.

```python
#lorentian function used for fit
def fitFunc(x,x0,gamma,norm):
    return 1/np.pi * gamma/( (x-x0)**2 +gamma**2 ) * norm
```

```python
#constructing binned data to use for fit
nBins=100
chi_range = (np.min(storeDict2["chi_eff"]), np.max(storeDict2["chi_eff"]))
binned_data, bins = np.histogram(storeDict2["chi_eff"], bins=nBins, range=chi_range)
centerBins = 0.5*(bins[1:] + bins[:-1])
#fit
pars, pars_cov = opt.curve_fit(fitFunc,centerBins,binned_data)
perr = np.sqrt(np.diag(pars_cov))
```

```python
#distribution of mass
%matplotlib inline
fig, ax = plt.subplots()
ax.grid()


values,bins,_ = ax.hist(storeDict2["chi_eff"], nBins, histtype='stepfilled', range=chi_range,
                           density=False, facecolor='Blue', alpha=0.75, label='Distribution')
_ = ax.set_title(r'Distribution of  $\chi_{eff}$ ')
_ = ax.set_xlabel(r'$\chi_{eff}$')
_ = ax.set_ylabel('Occurences')

#plot fit
_ = plt.plot(bins, fitFunc(bins,pars[0],pars[1],pars[2]),color='Red',lw=3,label='Best Fit')

_=ax.legend()

#Display text box with fitparameters outside plot
textstr = '\n'.join((
    'Fit parameters:',
    r'$x_{0}=%.2f \pm %.2f$' % (pars[0], perr[0] ),
    r'$\Gamma=%.2f \pm %.2f$' % (pars[1], perr[1], ),
    r'$Norm =%.2f \pm %.2f$' % (pars[2], perr[2], ), ))

props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
# place the text box in upper left in axes coords
_ = ax.text(1.05, 0.95, textstr, transform=ax.transAxes, fontsize=14,
        verticalalignment='top', bbox=props)
```
